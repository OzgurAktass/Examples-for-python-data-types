```python
###############################################
# TASK 5: Write a function that takes a list as argument, assigns the odd and even numbers in the list to separate lists and returns these lists.
###############################################


def ayir(l):
    tek = []
    cift = []
    for sayi in l:
        if sayi % 2 == 0:
            cift.append(sayi)
        else:
            tek.append(sayi)
    return tek, cift

l = [2, 13, 18, 93, 22]
tek, cift = ayir(l)
print("Tekler:", tek)
print("Çiftler:", cift)

```

```python
###############################################
# TASK 6: The list below contains the names of the students who ranked in the engineering and medical faculties.
# The first three students represent the ranking of engineering students and the last three students represent the ranking of medical students.
# Print the student rankings by faculty using Enumarate.
###############################################

ogrenciler = ["Ali","Veli","Ayşe","Talat","Zeynep","Ece"]

for siralama , ogrenci in enumerate(ogrenciler,1):
    if siralama < 4:
        print(f"Mühendislik fakültesi {siralama}. Öğrenci:{ogrenci}")
    else:
        print(f"Tıp fakültesi {siralama-3}. Öğrenci:{ogrenci}")
```

```python
###############################################
# TASK 7: 3 lists are given below. The lists include the code, credit and quota information of a course respectively. Print the course information using Zip.
###############################################
ders_kodu = ["CMP1005","PSY1001","HUK1005","SEN2204"]
kredi = [3,4,2,4]
kontenjan = [30,75,150,25]


for ders in zip(ders_kodu, kredi, kontenjan):
    print(f"Kredisi {ders[1]} olan {ders[0]} kodlu dersin kontenjanı {ders[2]} kişidir.")
```

```python
###############################################
# TASK 8: 2 sets are given below.
# You are expected to define a function that prints the common elements of set 1 if it covers set 2 and the difference of set 2 from set 1 if it does not.
###############################################

kume1 = set(["data", "python"])
kume2 = set(["data", "function", "qcut", "lambda", "python", "miuul"])

if kume1.issuperset(kume2):
    print(kume1.intersection(kume2))
else:
    print(kume2.difference(kume1))
```
